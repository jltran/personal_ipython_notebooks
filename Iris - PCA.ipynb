```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

iris = datasets.load_iris()
%matplotlib inline
```

```python
print iris.keys()
print iris.target_names
```

```python
labels = iris.feature_names
df = pd.DataFrame.from_records(iris.data, columns=labels)
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
X_std = StandardScaler().fit_transform(df)
X_std
```

```python
sklearn_pca = PCA(n_components = 2)
Y_sklearn = sklearn_pca.fit_transform(X_std)
Y_sklearn
```

```python
df = pd.DataFrame(Y_sklearn)
```

```python
df['label'] = iris.target
df
```

```python
colors = ['#1b9e77', '#d95f02', '#7570b3']
for i in xrange(len(iris.target_names)):
    plt.scatter(x=df[0][df.label==i], y=df[1][df.label==i] ,c=colors[i], s=30, alpha=0.75, label=iris.target_names[i])

plt.legend()
plt.show()
```
